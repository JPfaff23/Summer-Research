# Initial Data Cleaning

```python
import pandas as pd
from pathlib import Path

def clean_parquet(src_path, dst_path):
    """
    • Divides every S0_* and price by the row’s K
    • Renames the results to `S0_i/K` and `price/k`
    • Removes *_se, the raw S0_*, price, and K columns
    • Drops rows where price/k < 1e-6
    """
    df = pd.read_parquet(src_path)

    # 1) locate relevant columns
    s0_cols = [c for c in df.columns if c.lower().startswith("s0_")]
    se_cols = [c for c in df.columns if c.endswith("_se")]

    # 2) create scaled columns with the requested names
    for col in s0_cols:
        df[f"{col}/K"] = df[col] / df["K"]
    df["price/k"] = df["price"] / df["K"]

    # 3) filter out extremely small payoffs
    df = df[df["price/k"] >= 1e-6]

    # 4) drop the original, now-redundant columns
    df = df.drop(columns=s0_cols + ["price", "K"] + se_cols)

    # 5) save the cleaned file
    df.to_parquet(dst_path, index=False)
    print(f"✔ Saved cleaned file → {dst_path!s}")

# Apply to both datasets
clean_parquet("Train_New.parquet", "Train_Clean.parquet")
clean_parquet("Test_New.parquet",  "Test_Clean.parquet")
```

# Neural Network

```python
# Cell 1 — imports & device setup
import os, math, random, time, pathlib, re
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

from torchmetrics import R2Score

# Use inline plots
%matplotlib inline

os.environ.setdefault("NCCL_P2P_LEVEL", "0")
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Running on:", DEVICE)

torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Cell 2 — load & split cleaned Parquet into train/val/test TensorDatasets

# 1) Read cleaned Parquet files
#    (assumes you have already run clean_parquet on Train.parquet and Test100M.parquet)
train_df = pd.read_parquet("Train_clean.parquet", engine="pyarrow")
test_df  = pd.read_parquet("Test_clean.parquet", engine="pyarrow")

# 2) Split train_df → 80% train, 20% validation
X_full = train_df.drop(columns=["price/k"]).values.astype(np.float32)
y_full = train_df["price/k"].values.astype(np.float32)
X_tr_np, X_val_np, y_tr_np, y_val_np = train_test_split(
    X_full, y_full, test_size=0.01, random_state=42
)

# 3) Build TensorDatasets
train_ds = TensorDataset(torch.from_numpy(X_tr_np), torch.from_numpy(y_tr_np))
val_ds   = TensorDataset(torch.from_numpy(X_val_np), torch.from_numpy(y_val_np))
test_ds  = TensorDataset(
    torch.from_numpy(test_df.drop(columns=["price/k"]).values.astype(np.float32)),
    torch.from_numpy(test_df["price/k"].values.astype(np.float32))
)

print(f"train {len(train_ds):,} rows")
print(f"valid {len(val_ds):,} rows")
print(f" test {len(test_ds):,} rows")

```

```python
# Cell 3 — define the MLP (BasketNet) model

class BasketNet(nn.Module):
    def __init__(self, in_dim: int, width: int, layers: int):
        """
        in_dim = number of input features
        width  = nodes per hidden layer
        layers = number of hidden layers
        """
        super().__init__()
        blocks = [nn.Linear(in_dim, width), nn.ReLU()]
        for _ in range(layers - 1):
            blocks += [nn.Linear(width, width), nn.ReLU()]
        blocks.append(nn.Linear(width, 1))
        self.net = nn.Sequential(*blocks)

        # Xavier initialization
        for m in self.net:
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, x):
        return self.net(x)


def run_experiment_updates_per_update(
    width: int = 300,
    layers: int = 4,
    batch_size: int = 50_000,
    n_updates: int = 100_000,     # total mini‐batch updates
    lr: float = 1e-3,
    optimizer_name: str = "Adam",
    seed: int = 42,
    save_model: bool = False,      # ← new flag
    log_every: int = 1             # record every `log_every` updates
):
    """
    Train for exactly n_updates mini‐batches, recording
    train/val MSE at each update, then evaluate on test set.

    Args:
      width, layers:    model architecture
      batch_size:       mini‐batch size
      n_updates:        total optimizer steps (updates)
      lr:               learning rate
      optimizer_name:   "Adam" | "SGD" | "LBFGS"
      seed:             random seed
      save_model:       if True, save final weights to disk
      log_every:        record/compute validation every log_every steps
    """
    # 5.1) Re‐seed for reproducibility
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)

    # 5.2) Build DataLoaders
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=batch_size, shuffle=False, pin_memory=True)
    test_loader  = DataLoader(test_ds,  batch_size=batch_size, shuffle=False, pin_memory=True)

    # 5.3) Instantiate the model
    in_dim = train_ds[0][0].shape[0]
    model = BasketNet(in_dim, width, layers).to(DEVICE)
    if DEVICE.type == "cuda" and torch.cuda.device_count() > 1:
        model = nn.DataParallel(model)

    # 5.4) Loss & optimizer
    criterion = nn.MSELoss()
    if optimizer_name == "Adam":
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name == "SGD":
        optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)
    elif optimizer_name == "LBFGS":
        optimizer = optim.LBFGS(model.parameters(), lr=lr, max_iter=20)
    else:
        raise ValueError(f"Unknown optimizer: {optimizer_name}")

    # 5.5) Create an infinite iterator over shuffled train_loader
    def infinite_train_iter(loader):
        while True:
            for batch in loader:
                yield batch

    inf_train = infinite_train_iter(train_loader)

    # 5.6) Storage for train/val losses
    train_losses   = []
    valid_losses   = []
    steps_recorded = []

    # decide how often to print (every 10% of total updates)
    print_every = max(1, n_updates // 10)

    # 5.7) Main loop: exactly n_updates optimizer steps
    model.train()
    for step in range(1, n_updates + 1):
        # ----- 1) one optimizer step -----
        Xb, yb = next(inf_train)
        Xb, yb = Xb.to(DEVICE), yb.to(DEVICE).unsqueeze(1)
        optimizer.zero_grad(set_to_none=True)
        out       = model(Xb)
        loss      = criterion(out, yb)
        loss.backward()
        optimizer.step()
        loss_train = loss.item()

        # ----- 2) record & validate every log_every steps -----
        if step == 1 or step == n_updates or (step % log_every == 0):
            # validation pass
            model.eval()
            tot_val, cnt_val = 0.0, 0
            with torch.no_grad():
                for Xv, yv in val_loader:
                    Xv, yv = Xv.to(DEVICE), yv.to(DEVICE).unsqueeze(1)
                    ov = model(Xv)
                    tot_val += criterion(ov, yv).item() * Xv.size(0)
                    cnt_val += Xv.size(0)
            val_mse = tot_val / cnt_val

            # stash losses
            steps_recorded.append(step)
            train_losses.append(loss_train)
            valid_losses.append(val_mse)

            model.train()

            # ----- 3) print only at the first, last, or each 10% milestone -----
            if step == 1 or step == n_updates or (step % print_every == 0):
                print(f"[upd {step}/{n_updates}] "
                      f"train MSE={loss_train:.3e}  val MSE={val_mse:.3e}")

    # ---- 4) plot on the down-sampled grid ----
    train_min = np.minimum.accumulate(train_losses)
    valid_min = np.minimum.accumulate(valid_losses)

    plt.figure(figsize=(7, 4))
    plt.plot(steps_recorded, np.log10(train_min), label="train (cum-min)")
    plt.plot(steps_recorded, np.log10(valid_min), label="valid (cum-min)")
    plt.xlabel("update step")
    plt.ylabel("log10 MSE")
    plt.title(f"{layers}×{width} • bs={batch_size} • {optimizer_name}")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # 5.9) Final test‐set evaluation
    model.eval()
    all_preds, all_truths = [], []
    with torch.no_grad():
        for Xb, yb in test_loader:
            pr = model(Xb.to(DEVICE)).cpu().squeeze()
            all_preds.append(pr)
            all_truths.append(yb)
    preds  = torch.cat(all_preds)
    truths = torch.cat(all_truths)
    test_mse = criterion(preds.unsqueeze(1), truths.unsqueeze(1)).item()
    test_r2  = R2Score()(preds, truths).cpu().item()
    print(f"\nTest results → MSE = {test_mse:.3e}   R² = {test_r2:.4f}")

    # 5.10) Optionally save model weights
    if save_model:
        tag = f"w{width}_L{layers}_bs{batch_size}_upd{n_updates}_" \
              f"{optimizer_name}_lr{lr:g}"
        torch.save(model.state_dict(), f"model_{tag}.pt")
        print(f"Saved model to model_{tag}.pt")
```

# Applying Regularization to Improve Validation MSE

```python
#!/usr/bin/env python3
import random, time
import numpy as np, torch
import torch.nn as nn, torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from torchmetrics import R2Score
import matplotlib.pyplot as plt

# Assuming train_ds, val_ds, test_ds, DEVICE are already defined globally

class BasketNet(nn.Module):
    def __init__(self, in_dim, width, layers,
                 dropout_p=0.0, n_drop_layers=2):
        super().__init__()
        blocks = [nn.Linear(in_dim, width), nn.GELU()]
        for i in range(layers - 1):
            blocks += [nn.Linear(width, width), nn.GELU()]
            if dropout_p > 0 and i >= layers - 1 - n_drop_layers:
                blocks.append(nn.Dropout(dropout_p))
        blocks.append(nn.Linear(width, 1))
        self.net = nn.Sequential(*blocks)
        for m in self.net:
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, x):
        return self.net(x)

def cosine_scheduler(optimizer, total_updates, eta_min=1e-5):
    return optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=total_updates, eta_min=eta_min)

def run_experiment_regularization(
    width: int = 300,
    layers: int = 4,
    batch_size: int = 50_000,
    n_updates: int = 100_000,
    lr: float = 1e-3,
    optimizer_name: str = "Adam",
    seed: int = 42,
    save_model: bool = False,
    log_every: int = 1,
    # new args ↓
    dropout_p: float = 0.0,
    weight_decay: float = 0.0,
    sched_name: str = None,          # "cosine" or "step"
    early_patience: int = None       # number of consecutive val‐checks
):
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)

    train_loader = DataLoader(train_ds, batch_size=batch_size,
                              shuffle=True,  pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=batch_size,
                              shuffle=False, pin_memory=True)
    test_loader  = DataLoader(test_ds,  batch_size=batch_size,
                              shuffle=False, pin_memory=True)

    in_dim = train_ds[0][0].shape[0]
    model = BasketNet(in_dim, width, layers,
                      dropout_p=dropout_p).to(DEVICE)
    if DEVICE.type=="cuda" and torch.cuda.device_count()>1:
        model = nn.DataParallel(model)

    # choose optimizer
    if optimizer_name=="Adam":
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name=="AdamW":
        optimizer = optim.AdamW(model.parameters(),
                                lr=lr, weight_decay=weight_decay)
    elif optimizer_name=="SGD":
        optimizer = optim.SGD(model.parameters(), lr=lr,
                              momentum=0.9, weight_decay=weight_decay)
    else:
        raise ValueError(f"Unknown optimizer: {optimizer_name}")

    # choose scheduler
    scheduler = None
    if sched_name=="cosine":
        scheduler = cosine_scheduler(optimizer, n_updates)
    elif sched_name=="step":
        scheduler = optim.lr_scheduler.StepLR(
            optimizer, step_size=n_updates//3, gamma=0.1)

    criterion = nn.MSELoss()
    r2_metric = R2Score().to(DEVICE)

    best_val    = float('inf')
    best_state  = None
    stale_count = 0

    train_hist, val_hist, steps = [], [], []

    inf_iter = iter(train_loader)
    print_every = max(1, n_updates//10)

    model.train()
    for step in range(1, n_updates+1):
        # --- one train step
        try:
            xb, yb = next(inf_iter)
        except StopIteration:
            inf_iter = iter(train_loader)
            xb, yb = next(inf_iter)

        xb = xb.to(DEVICE); yb = yb.to(DEVICE).unsqueeze(1)
        optimizer.zero_grad(set_to_none=True)
        out = model(xb)
        loss = criterion(out, yb)
        loss.backward()
        optimizer.step()
        if scheduler: scheduler.step()

        # --- validation & logging
        if step % log_every == 0 or step in (1, n_updates):
            model.eval()
            tot, cnt = 0.0, 0
            with torch.no_grad():
                for xv, yv in val_loader:
                    xv = xv.to(DEVICE); yv = yv.to(DEVICE).unsqueeze(1)
                    tot += criterion(model(xv), yv).item() * len(xv)
                    cnt += len(xv)
            val_mse = tot / cnt

            steps.append(step)
            train_hist.append(loss.item())
            val_hist.append(val_mse)

            # early stopping logic
            if val_mse < best_val - 1e-12:
                best_val  = val_mse
                best_state= model.state_dict()
                stale_count = 0
            else:
                stale_count += 1

            if step % print_every == 0 or step in (1, n_updates):
                print(f"[{step}/{n_updates}] "
                      f"train={loss.item():.3e} val={val_mse:.3e} "
                      f"LR={optimizer.param_groups[0]['lr']:.2e}")

            model.train()
            if early_patience and stale_count >= early_patience:
                print(f"Early stopping after {stale_count} val checks.")
                break

    # restore best
    if best_state:
        model.load_state_dict(best_state)

    # plot
    plt.figure(figsize=(7,4))
    plt.plot(steps, np.log10(np.minimum.accumulate(train_hist)),
             label="train (cum-min)")
    plt.plot(steps, np.log10(np.minimum.accumulate(val_hist)),
             label="valid (cum-min)")
    plt.xlabel("update"); plt.ylabel("log10 MSE")
    plt.title(f"{layers}×{width} • bs={batch_size} • opt={optimizer_name}")
    plt.legend(); plt.show()

    # final test
    model.eval()
    preds, truths = [], []
    with torch.no_grad():
        for xb, yb in test_loader:
            preds.append(model(xb.to(DEVICE)).cpu().squeeze())
            truths.append(yb)
    preds  = torch.cat(preds)
    truths = torch.cat(truths)
    test_mse = criterion(preds.unsqueeze(1), truths.unsqueeze(1)).item()
    test_r2  = r2_metric(preds, truths).item()
    print(f"\nTest → MSE={test_mse:.3e}   R²={test_r2:.4f}")

    # save if desired
    if save_model:
        tag = f"W{width}_L{layers}_bs{batch_size}_upd{step}_" \
              f"{optimizer_name}_lr{lr:g}" + \
              (f"_wd{weight_decay}" if weight_decay else "") + \
              (f"_dop{dropout_p}"     if dropout_p    else "")
        torch.save(model.state_dict(), f"model_{tag}.pt")
        print("Saved model:", tag)
```

```python
#!/usr/bin/env python3
import random, argparse, time
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import OneCycleLR, StepLR, CosineAnnealingLR
from torch.optim.swa_utils import AveragedModel, SWALR, update_bn
from torch.utils.data import DataLoader, TensorDataset
from torchmetrics import R2Score

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.benchmark = True

# ----------------------- model ----------------------------- #
class BasketNet(nn.Module):
    def __init__(self, in_dim, width, layers,
                 dropout_p=0.0, use_layernorm=False,
                 n_drop_layers=2):
        super().__init__()
        blocks = [nn.Linear(in_dim, width),
                  nn.LayerNorm(width) if use_layernorm else nn.Identity(),
                  nn.GELU()]
        for i in range(layers - 1):
            blocks += [nn.Linear(width, width),
                       nn.LayerNorm(width) if use_layernorm else nn.Identity(),
                       nn.GELU()]
            if dropout_p and i >= layers - 1 - n_drop_layers:
                blocks.append(nn.Dropout(dropout_p))
        blocks.append(nn.Linear(width, 1))
        self.net = nn.Sequential(*blocks)
        for m in self.net:
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, x): return self.net(x)

# -------------------- experiment --------------------------- #
def run_experiment_regularization_2(
        width=300, layers=4, batch_size=50_000,
        n_updates=100_000, lr=1e-3, optimizer_name="AdamW",
        dropout_p=0.0, weight_decay=0.0,
        sched="cosine", onecycle=False, swa=False,  # new flags
        seed=42, log_every=500, early_patience=None,
        use_layernorm=False, save_model=False):

    torch.manual_seed(seed); np.random.seed(seed); random.seed(seed)

    # ---- load data (expects Train_clean.parquet / Test_clean.parquet)
    train_df = pd.read_parquet("Train_clean.parquet")
    test_df  = pd.read_parquet("Test_clean.parquet")

    X = train_df.drop(columns=["price/k"]).values.astype(np.float32)
    y = train_df["price/k"].values.astype(np.float32)
    X_tr, X_val, y_tr, y_val = train_test_split(
        X, y, test_size=0.01, shuffle=True, random_state=seed)

    def ds(a, b): return TensorDataset(torch.from_numpy(a), torch.from_numpy(b))
    train_ds, val_ds = ds(X_tr, y_tr), ds(X_val, y_val)
    test_ds  = ds(test_df.drop(columns=["price/k"]).values.astype(np.float32),
                  test_df["price/k"].values.astype(np.float32))

    train_loader = DataLoader(train_ds, batch_size=batch_size,
                              shuffle=True, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=batch_size,
                              shuffle=False, pin_memory=True)
    test_loader  = DataLoader(test_ds,  batch_size=batch_size,
                              shuffle=False, pin_memory=True)

    in_dim = X.shape[1]
    model  = BasketNet(in_dim, width, layers,
                       dropout_p, use_layernorm).to(DEVICE)
    if DEVICE.type == "cuda" and torch.cuda.device_count() > 1:
        model = nn.DataParallel(model)

    # ---- optimiser
    if optimizer_name == "Adam":
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name == "AdamW":
        optimizer = optim.AdamW(model.parameters(), lr=lr,
                                weight_decay=weight_decay)
    elif optimizer_name == "SGD":
        optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9,
                              weight_decay=weight_decay)
    else:
        raise ValueError("opt must be Adam / AdamW / SGD")

    # ---- scheduler
    if onecycle:
        scheduler = OneCycleLR(optimizer, max_lr=lr, total_steps=n_updates,
                               pct_start=0.1, div_factor=10,
                               final_div_factor=1e4)
    elif sched == "step":
        scheduler = StepLR(optimizer, step_size=n_updates // 3, gamma=0.1)
    else:   # default cosine
        scheduler = CosineAnnealingLR(optimizer, T_max=n_updates)

    # ---- SWA wrapper
    if swa:
        swa_start = int(0.6 * n_updates)
        swa_model = AveragedModel(model)
        swa_sched = SWALR(optimizer, swa_lr=lr * 0.1)

    criterion = nn.MSELoss()
    r2_metric = R2Score().to(DEVICE)

    best_val, best_state, stale = float("inf"), None, 0
    tr_hist = []; va_hist = []; steps = []

    train_iter = iter(train_loader)
    print_every = max(1, n_updates // 10)

    for step in range(1, n_updates + 1):
        try:
            xb, yb = next(train_iter)
        except StopIteration:
            train_iter = iter(train_loader)
            xb, yb = next(train_iter)

        xb, yb = xb.to(DEVICE), yb.to(DEVICE).unsqueeze(1)
        optimizer.zero_grad(set_to_none=True)
        loss = criterion(model(xb), yb)
        loss.backward(); optimizer.step()

        if swa and step >= swa_start:
            swa_model.update_parameters(model)
            swa_sched.step()
        else:
            scheduler.step()

        # ---- validation
        if step % log_every == 0 or step in (1, n_updates):
            with torch.no_grad():
                model.eval(); tot, cnt = 0., 0
                for xv, yv in val_loader:
                    xv, yv = xv.to(DEVICE), yv.to(DEVICE).unsqueeze(1)
                    tot += criterion(model(xv), yv).item() * len(xv)
                    cnt += len(xv)
            val_mse = tot / cnt
            tr_hist.append(loss.item()); va_hist.append(val_mse); steps.append(step)

            if val_mse < best_val - 1e-12:
                best_val, best_state, stale = val_mse, model.state_dict(), 0
            else:
                stale += 1

            if step % print_every == 0 or step in (1, n_updates):
                print(f"[{step}/{n_updates}] "
                      f"train={loss.item():.3e} val={val_mse:.3e} "
                      f"LR={optimizer.param_groups[0]['lr']:.2e}")
            model.train()
            if early_patience and stale >= early_patience:
                print("Early stop."); break

    # ---- restore best & (optional) SWA bn update
    model.load_state_dict(best_state)
    if swa:
        update_bn(train_loader, swa_model)
        model = swa_model

    # ---- plot
    plt.figure(figsize=(7,4))
    plt.plot(steps, np.log10(np.minimum.accumulate(tr_hist)), label="train (cum-min)")
    plt.plot(steps, np.log10(np.minimum.accumulate(va_hist)), label="valid (cum-min)")
    plt.xlabel("update"); plt.ylabel("log10 MSE")
    plt.title(f"{layers}×{width} bs={batch_size} opt={optimizer_name}"
              + (" +SWA" if swa else "") + (" +LN" if use_layernorm else ""))
    plt.legend(); plt.tight_layout(); plt.show()

    # ---- final test
    model.eval(); preds, truths = [], []
    with torch.no_grad():
        for xb, yb in test_loader:
            preds.append(model(xb.to(DEVICE)).cpu().squeeze())
            truths.append(yb)
    preds = torch.cat(preds); truths = torch.cat(truths)
    test_mse = criterion(preds, truths).item()
    test_r2  = r2_metric(preds.to(DEVICE), truths.to(DEVICE)).cpu().item()
    print(f"\nTest → MSE={test_mse:.3e}  R²={test_r2:.4f}")

    if save_model:
        tag = f"L{layers}_W{width}_bs{batch_size}_{optimizer_name}" \
              f"_up{n_updates}{'_oc' if onecycle else ''}" \
              f"{'_swa' if swa else ''}"
        torch.save(model.state_dict(), f"model_{tag}.pt")
        print("Saved:", tag)

```
